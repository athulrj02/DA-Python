```python
# 4. Create a single histogram-plot containing all dimensions
# data = load_iris()
df = pd.DataFrame(data=data.data, columns=data.feature_names)

plt.figure(figsize=(10, 6))

for feature in df.columns:
    plt.hist(df[feature], bins=20, alpha=0.5, label=feature)

plt.title('Combined Histogram for Iris Dataset Features')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend(loc='upper right')

plt.show()
```

```python
# 5. Create a boxplot over the four dimensions
plt.figure(figsize = (10,6))
plt.boxplot(df)
plt.xticks([1,2,3,4], feature_names, rotation=45)
plt.title('Iris Boxplot')
plt.show()
```
